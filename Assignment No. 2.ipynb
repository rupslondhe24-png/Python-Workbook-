<h1>Question Set A</h1>

```python
# Program to count the number of vowels in a string
S = input("Enter a word or a sentence: ")
c = 0
for i in S.lower():
    if i in "aeiou": # Check if the character is a vowel
        c += 1 # Increment the vowel count
print(f"Total vowels: {c}")
```

```python
# Program to check palindrome strings
S = input("Enter a word or a sentence: ")
if S[::-1] == S: # Check if the string is the same when reversed
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Program to reverse a string without using built-in reverse functions
S = input("Enter a word or a sentence: ")
R = "" # Initialize an empty string for the reversed result
for ch in S:
    R = ch + R # Append each character in reverse order
print(R)
```

```python
# Program to count the frequency of each character in a string
S = input("Enter a word or a sentence: ")
c = 0
D = dict() # Create an empty dictionary to store character counts
for s in S:
    if s in D:
        D[s] += 1 # Increment count if character already in dictionary
    else:
        D[s] = 1 # Initialize count to 1 if character not in dictionary
for k in D:
    print(f"'{k}': {D[k]}") # Print each character and its count 
```

```python
# Program to remove all whitespaces from a string
S = input("Enter a word or a sentence: ")
R = S.replace(" ", "") # Remove all spaces
print(R)
```

<h1>Question Set B</h1>

```python
# Program to find the first non-repeating character in a string
S = input("Enter a word or a sentence: ")
for s in S:
    if S.count(s) == 1: # Check if the character appears only once
        print("First non-repeating character is:", s)
        break # Exit the loop after finding the first non-repeating character
```

```python
# Program to check if two strings are anagrams
s1 = input("Enter first string: ")
s2 = input("Enter second string: ")
if sorted(s1) == sorted(s2): # Sort and compare the two strings
    print("Anagrams") # If sorted strings are equal, they are anagrams
else:
    print("Not Anagrams") # If not equal, they are not anagrams
```

```python
# Program to convert a string to a title case without using title() method
s = input("Enter a word or a sentence: ")
result = ""
capitalize_next = True
for char in s:
    if char == ' ': # If the character is a space
        result += char # Keep the space
        capitalize_next = True
    elif capitalize_next: # If the next character should be capitalized
        result += char.upper() # Capitalize the character
        capitalize_next = False
    else: # If the character should not be capitalized
        result += char # Keep the character as is
print(result)
```

```python
# Program to replace all occurences of a substring with another one in a string
s = input("Enter a word or a sentence: ")
old_substring = input("Enter substring to replace it with: ")
new_substring = input("Enter the new substring: ")
result = ""
i = 0
while i < len(s): # Loop through the string s
    if s[i:i+len(old_substring)] == old_substring: # Check for old_substring
        result += new_substring # Replace with new_substring
        i += len(old_substring) # Move index forward by length of old_substring
    else:
        result += s[i] # Keep the current character
        i += 1 # Move to the next character
print(result)
```

<h1>Question Set C</h1>

```python
# Program to find the longest word in a string
S = input("Enter a string: ")
L = S.split() # Split the string into words
longest_word = "" # Initialize variable to store the longest word
for word in L:
    if len(word) > len(longest_word): # Compare lengths of words
        longest_word = word # Update longest_word if current word is longer
print("The longest word is:", longest_word) # Display the longest word
```

```python
# Program to see if a string contains only numbers, alphabets, or both
S = input("Enter a string: ")
if S.isdigit():
    print("Has only numbers")
elif S.isalpha():
    print("Has only alphabets")
elif S.isalnum():
    print("Has both numbers and alphabets")
else:
    print("None are present")
```
