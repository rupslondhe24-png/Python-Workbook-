<h1>Question Set A</h1>

```python
# Function to find maximum of three numbers
def max_of_three(a, b, c):
    max_val = a
    if b > max_val:
        max_val = b
    if c > max_val:
        max_val = c
    return max_val

x = int(input("Enter first: "))
y = int(input("Enter second: "))
z = int(input("Enter third: "))
print("Maximum is:", max_of_three(x, y, z))
```

```python
# Function to check if number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

num = int(input("Enter number: "))
if is_prime(num):
    print("Prime")
else:
    print("Not prime")
```

```python
# Iterative factorial
def fact_iter(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

# Recursive factorial
def fact_rec(n):
    if n == 0 or n == 1:
        return 1                            # base case
    return n * fact_rec(n - 1)

n = int(input("Enter n: "))
print("Iterative factorial:", fact_iter(n))
print("Recursive factorial:", fact_rec(n))
```

```python
# Function that returns list of unique elements
def unique_list(lst):
    seen = []
    for x in lst:
        if x not in seen:
            seen.append(x)
    return seen

data = list(map(int, input("Enter numbers: ").split()))
print("Unique elements:", unique_list(data))
```

```python
# Function to calculate sum of all even numbers in a list
def sum_even(lst):
    total = 0
    for x in lst:
        if x % 2 == 0:
            total += x
    return total

nums = list(map(int, input("Enter numbers: ").split()))
print("Sum of even numbers:", sum_even(nums))
```

<h1>Question Set B</h1>

```python
# Functions to convert temperature
def c_to_f(c):
    return (c * 9/5) + 32                    # Celsius to Fahrenheit

def f_to_c(f):
    return (f - 32) * 5/9                    # Fahrenheit to Celsius

choice = input("Convert from (C/F)? ").upper()
temp = float(input("Enter temperature: "))

if choice == "C":
    print("In Fahrenheit:", c_to_f(temp))
elif choice == "F":
    print("In Celsius:", f_to_c(temp))
else:
    print("Invalid choice")
```

```python
# Function that returns count of vowels and consonants
def count_vowels_consonants(s):
    vowels_set = "aeiouAEIOU"
    v = c = 0
    for ch in s:
        if ch.isalpha():                     # only letters
            if ch in vowels_set:
                v += 1
            else:
                c += 1
    return v, c

text = input("Enter a string: ")
vowels, consonants = count_vowels_consonants(text)
print("Vowels:", vowels)
print("Consonants:", consonants)
```

```python
# Function that accepts variable number of arguments and returns average
def average(*nums):
    if len(nums) == 0:
        return 0
    total = 0
    for x in nums:
        total += x
    return total / len(nums)

print("Average:", average(1, 2, 3, 4, 5))
```

```python
# Lambda function to find the maximum of three numbers
max_three = lambda a, b, c: a if (a >= b and a >= c) else (b if b >= c else c)

x = int(input("Enter first: "))
y = int(input("Enter second: "))
z = int(input("Enter third: "))

print("Maximum is:", max_three(x, y, z))
```

<h1>Question Set C</h1>

```python
# Recursive function to calculate sum of digits of a number
def sum_of_digits(n):
    n = abs(n)                                # handle negative numbers
    if n < 10:
        return n                              # base case: single digit
    return (n % 10) + sum_of_digits(n // 10)  # last digit + recursion

num = int(input("Enter a number: "))
print("Sum of digits:", sum_of_digits(num))
```

```python
# Function that returns dictionary with count of even and odd numbers
def count_even_odd(lst):
    counts = {"even": 0, "odd": 0}
    for x in lst:
        if x % 2 == 0:
            counts["even"] += 1
        else:
            counts["odd"] += 1
    return counts

nums = list(map(int, input("Enter numbers: ").split()))
result = count_even_odd(nums)
print(result)                                 
```
