<h1>Question Set A</h1>

```python
# Program to create dictionary roll -> name for 5 students
students = {}

for i in range(5):
    roll = int(input("Enter roll number: "))
    name = input("Enter name: ")
    students[roll] = name                   # key = roll

print("Student dictionary:")
for r, n in students.items():
    print(r, ":", n)
```

```python
# Program to count frequency of each word in a sentence
sentence = input("Enter a sentence: ")
words = sentence.split()

freq = {}
for w in words:
    freq[w] = freq.get(w, 0) + 1            # count words

print("Word frequencies:")
for w, c in freq.items():
    print(w, ":", c)
```

```python
# Program to merge two dictionaries
d1 = {}
d2 = {}

n1 = int(input("How many items in dict1? "))
for i in range(n1):
    key = input("Enter key for dict1: ")
    val = input("Enter value: ")
    d1[key] = val

n2 = int(input("How many items in dict2? "))
for i in range(n2):
    key = input("Enter key for dict2: ")
    val = input("Enter value: ")
    d2[key] = val

merged = d1.copy()                          # copy d1 first
merged.update(d2)                           # then add d2

print("Merged dictionary:", merged)
```

```python
# Program to find key with maximum value in a dictionary
d = {"a": 10, "b": 25, "c": 15}

max_key = None
max_val = None

for k, v in d.items():
    if max_val is None or v > max_val:
        max_val, max_key = v, k

print("Key with maximum value:", max_key)
```

```python
# Program to check if a given key exists in a dictionary
d = {"a": 1, "b": 2, "c": 3}
key = input("Enter key to search: ")

if key in d:                                 # membership on keys
    print("Key exists; value =", d[key])
else:
    print("Key does not exist")
```

<h1>Question Set B</h1>

```python
# Program to remove a key from dictionary and handle missing key
d = {"a": 1, "b": 2, "c": 3}
key = input("Enter key to remove: ")

if key in d:
    value = d.pop(key)                       # remove and return value
    print("Removed:", key, "value:", value)
else:
    print("Key not found; nothing removed")

print("Dictionary now:", d)
```

```python
# Program to sort dictionary by values in descending order
d = {"a": 5, "b": 2, "c": 8, "d": 1}

sorted_items = sorted(d.items(), key=lambda x: x[1], reverse=True)
sorted_dict = {}

for k, v in sorted_items:
    sorted_dict[k] = v                       # rebuild ordering

print("Sorted dict:", sorted_dict)
```

```python
# Program to create dictionary from two lists
keys = input("Enter keys separated by space: ").split()
values = input("Enter values separated by space: ").split()

d = {}
for i in range(min(len(keys), len(values))):
    d[keys[i]] = values[i]                   # map key[i] -> value[i]

print("Resulting dictionary:", d)
```

<h1>Question Set C</h1>

```python
# Program to find all keys with the same value in a dictionary
d = {"a": 1, "b": 2, "c": 1, "d": 3, "e": 2}

value_to_keys = {}                            # value -> list of keys
for k, v in d.items():
    if v not in value_to_keys:
        value_to_keys[v] = []                # create list for this value
    value_to_keys[v].append(k)

for v, keys in value_to_keys.items():
    if len(keys) > 1:                        # values shared by multiple keys
        print("Value", v, "has keys:", keys)
```

```python
# Program to create dictionary of cubes from 1 to n
n = int(input("Enter n: "))

cubes = {}
for i in range(1, n + 1):
    cubes[i] = i ** 3                        # key = i, value = i^3

print("Cubes dictionary:", cubes)
```
