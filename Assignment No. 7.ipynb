```python
# Class Rectangle with area and perimeter
class Rectangle:
    def __init__(self, length, width):
        self.length = length                 # instance variables
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

l = float(input("Enter length: "))
w = float(input("Enter width: "))
r = Rectangle(l, w)
print("Area:", r.area())
print("Perimeter:", r.perimeter())
```

```python
# Class BankAccount with deposit, withdraw, display
class BankAccount:
    def __init__(self, owner, balance=0.0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print("Deposited:", amount)

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print("Withdrawn:", amount)
        else:
            print("Insufficient balance")

    def display(self):
        print("Owner:", self.owner, "Balance:", self.balance)

acc = BankAccount("John", 1000)
acc.deposit(500)
acc.withdraw(300)
acc.display()
```

```python
# Parent class Vehicle and child classes Car and Bike
class Vehicle:
    def __init__(self, name):
        self.name = name

    def move(self):
        print(self.name, "is moving")

class Car(Vehicle):
    def move(self):
        print(self.name, "is driving on road")   # overridden

class Bike(Vehicle):
    def move(self):
        print(self.name, "is riding on road")    # overridden

c = Car("Car1")
b = Bike("Bike1")
c.move()
b.move()
```

```python
# Class Student with private attributes and public access methods
class Student:
    def __init__(self, name, roll, marks):
        self.__name = name                      # private
        self.__roll = roll
        self.__marks = marks

    def get_name(self):
        return self.__name

    def get_roll(self):
        return self.__roll

    def get_marks(self):
        return self.__marks

    def set_marks(self, marks):
        if 0 <= marks <= 100:
            self.__marks = marks

s = Student("John", 101, 85)
print("Name:", s.get_name())
print("Roll:", s.get_roll())
print("Marks:", s.get_marks())
```

```python
# Class Circle with area and overloaded + operator
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * (self.radius ** 2)

    def __add__(self, other):
        # returns new Circle whose radius is sum of radii
        return Circle(self.radius + other.radius)

c1 = Circle(3)
c2 = Circle(4)
c3 = c1 + c2                                 # uses __add__
print("New circle radius:", c3.radius)
print("New circle area:", c3.area())
```

```python
# Multilevel inheritance: Animal -> Mammal -> Dog
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(self.name, "is eating")

class Mammal(Animal):
    def walk(self):
        print(self.name, "is walking")

class Dog(Mammal):
    def bark(self):
        print(self.name, "is barking")

d = Dog("Tommy")
d.eat()                                      # from Animal
d.walk()                                     # from Mammal
d.bark()                                     # from Dog
```

```python
# Employee class with class method to count total employees
class Employee:
    count = 0                                # class attribute

    def __init__(self, name):
        self.name = name
        Employee.count += 1                  # increment when object created

    @classmethod
    def get_count(cls):
        return cls.count

e1 = Employee("A")
e2 = Employee("B")
e3 = Employee("C")
print("Total employees:", Employee.get_count())
```

```python
# Book class with special methods
class Book:
    def __init__(self, title, pages):
        self.title = title
        self.pages = pages

    def __str__(self):
        return f"{self.title} ({self.pages} pages)"  # nice string representation

    def __len__(self):
        return self.pages                            # len(book) -> pages

    def __eq__(self, other):
        return self.pages == other.pages             # compare by pages

book1 = Book("Python Basics", 200)
book2 = Book("Advanced Python", 300)
book3 = Book("Another Book", 200)

print(book1)                     # uses __str__
print(len(book1))                # uses __len__
print(book1 == book2)            # False
print(book1 == book3)            # True
```

```python
# Temperature class with Celsius and Fahrenheit properties
class Temperature:
    def __init__(self):
        self._celsius = 0

    @property
    def celsius(self):
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        if value < -273.15:                         # physical lower limit
            raise ValueError("Temperature below absolute zero")
        self._celsius = value

    @property
    def fahrenheit(self):
        return (self._celsius * 9/5) + 32

temp = Temperature()
temp.celsius = 25
print("Celsius:", temp.celsius)
print("Fahrenheit:", temp.fahrenheit)
```

```python
# Abstract class Shape and concrete subclasses
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass                                   # must be overridden

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * (self.radius ** 2)

class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

r = Rectangle(5, 3)
c = Circle(4)
t = Triangle(6, 2)

print("Rectangle area:", r.area())
print("Circle area:", c.area())
print("Triangle area:", t.area())
```

```python
# Library class to manage collection of books
class Library:
    def __init__(self):
        self.books = []                        # list of book titles (simple)

    def add_book(self, title):
        self.books.append(title)
        print("Added book:", title)

    def remove_book(self, title):
        if title in self.books:
            self.books.remove(title)
            print("Removed book:", title)
        else:
            print("Book not found")

    def search_book(self, title):
        if title in self.books:
            print("Book is available")
        else:
            print("Book is not available")

    def display_books(self):
        if not self.books:
            print("No books in library")
        else:
            print("Books in library:")
            for b in self.books:
                print("-", b)

lib = Library()
lib.add_book("Python Basics")
lib.add_book("Data Science 101")
lib.display_books()
lib.search_book("Python Basics")
lib.remove_book("Python Basics")
lib.display_books()
```
