<h1>Question Set A</h1>

```python
# Program to find the maximum and minimum element in a tuple without using max() and min() functions.
my_tuple = (5, 2, 8, 1, 9)
if not my_tuple:  # Check if tuple is empty to avoid errors
    print("Tuple is empty")
else:
    min_val = my_tuple[0]  # Initialize min with first element
    max_val = my_tuple[0]  # Initialize max with first element
    for num in my_tuple:  # Iterate through each element
        if num < min_val:  # Compare to find smaller value
            min_val = num  # Update minimum if smaller found
        if num > max_val:  # Compare to find larger value
            max_val = num  # Update maximum if larger found
    print(f"Minimum: {min_val}")
    print(f"Maximum: {max_val}")
```

```python
# Program to count the occurrence of each element in a tuple.
my_tuple = (1, 2, 3, 2, 1, 3, 2)
counts = {}  # Dictionary to store element:count pairs
for num in my_tuple:  # Loop through each element
    if num in counts:  # Check if element already counted
        counts[num] += 1  # Increment existing count
    else:  # First time seeing this element
        counts[num] = 1  # Initialize count to 1
for num, count in counts.items():
    print(f"{num}:{count}")
```

```python
# Program to convert a tuple of strings into a single string.
my_tuple = ("Hello", "World", "Python")
result = ""
for s in my_tuple:  # Iterate through string elements
    result += s  # Concatenate each string (no spaces)
print(result)
```

```python
# Program to find the sum of all even numbers in a tuple.
my_tuple = (1, 2, 3, 4, 5, 6, 7, 8)
total = 0  # Initialize sum accumulator
for num in my_tuple:  # Check each number
    if num % 2 == 0:  # Modulo 2 == 0 means even number
        total += num  # Add even number to running total
print(f"Sum of even numbers: {total}")
```

```python
# Program to remove duplicates from a tuple and return another with unique elements
t = (1, 2, 3, 2, 1, 4, 3)
seen = []

for x in t:
    if x not in seen:  # preserve first occurrence order
        seen.append(x)

result = tuple(seen)
print("Tuple without duplicates:", result)
```

<h1>Question Set B</h1>

```python
# Program to find second largest element in a tuple
t = (5, 1, 9, 3, 7)

first = second = None

for x in t:
    if first is None or x > first:
        second = first                     # update second before first
        first = x
    elif (second is None or x > second) and x != first:
        second = x

print("Second largest:", second)
```

```python
# Program to sort a tuple of tuples based on the second element
t = ((1, 5), (3, 2), (2, 8))

sorted_list = sorted(t, key=lambda x: x[1]) # key uses second element
result = tuple(sorted_list)

print("Sorted:", result)
```

```python
# Program to check if all elements in a tuple are unique
t = (1, 2, 3, 4, 5)

if len(t) == len(set(t)):           # set removes duplicates
    print("All elements are unique")
else:
    print("Elements are not unique")
```

```python
# Program to swap first and last element of a tuple
t = (1, 2, 3, 4, 5)

if len(t) >= 2:
    first = t[0]
    last = t[-1]
    middle = t[1:-1]

    new_t = (last,) + middle + (first,)    # note commas to keep tuples
else:
    new_t = t

print("After swap:", new_t)
```

<h1>Question Set C</h1>

```python
# Program to find common elements between two tuples
t1 = tuple(map(int, input("Enter elements of first tuple: ").split()))
t2 = tuple(map(int, input("Enter elements of second tuple: ").split()))

set1 = set(t1)
set2 = set(t2)

common = set1 & set2                        # intersection
result = tuple(common)                      # convert back to tuple

print("Common elements:", result)
```

```python
# Program to create tuple of squares from 1 to n
n = int(input("Enter n: "))

squares = []
for i in range(1, n + 1):
    squares.append(i * i)

result = tuple(squares)
print("Squares tuple:", result)
```
