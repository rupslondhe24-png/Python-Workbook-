<h1>Question Set A</h1>

```python
# Program for union, intersection, and difference of two sets
set1 = set(map(int, input("Enter elements of set1: ").split()))
set2 = set(map(int, input("Enter elements of set2: ").split()))

union_set = set1 | set2                     # union
inter_set = set1 & set2                     # intersection
diff_set = set1 - set2                      # only in set1

print("Union:", union_set)
print("Intersection:", inter_set)
print("Difference (set1 - set2):", diff_set)
```

```python
# Program to check if a set is subset of another
A = set(map(int, input("Enter elements of set A: ").split()))
B = set(map(int, input("Enter elements of set B: ").split()))

if A.issubset(B):                            # same as A <= B
    print("A is subset of B")
else:
    print("A is not subset of B")
```

```python
# Program to remove duplicates from list using set
lst = list(map(int, input("Enter list elements: ").split()))

unique_set = set(lst)                        # drops duplicates
unique_list = list(unique_set)

print("List without duplicates:", unique_list)
```

```python
# Program to find elements present in first set but not second
A = set(map(int, input("Enter elements of set A: ").split()))
B = set(map(int, input("Enter elements of set B: ").split()))

only_in_A = A - B                            # difference

print("Elements only in A:", only_in_A)
```

```python
# Program to check if two sets are disjoint
A = set(map(int, input("Enter elements of set A: ").split()))
B = set(map(int, input("Enter elements of set B: ").split()))

if A.isdisjoint(B):                          # no common elements
    print("Sets are disjoint")
else:
    print("Sets are not disjoint")
```

<h1>Question Set B</h1>

```python
# Program to find symmetric difference between two sets
A = set(map(int, input("Enter elements of set A: ").split()))
B = set(map(int, input("Enter elements of set B: ").split()))

sym_diff = A ^ B                             # elements in A or B but not both
print("Symmetric difference:", sym_diff)
```

```python
# Program to add multiple elements to a set from user
s = set()
n = int(input("How many elements to add? "))

to_add = []
for i in range(n):
    val = input("Enter element: ")
    to_add.append(val)                   # collect all input first

s.update(to_add)                         # update adds all elements
print("Final set:", s)
```

```python
# Program to find length of set and check if element exists
s = set(map(int, input("Enter elements of set: ").split()))

print("Length of set:", len(s))

x = int(input("Enter element to search: "))

if x in s:
    print(x, "exists in set")
else:
    print(x, "does not exist in set")
```

<h1>Question Set C</h1>

```python
# Program on two sentences: common and unique words
s1 = input("Enter first sentence: ")
s2 = input("Enter second sentence: ")

words1 = set(s1.split())                     # split into words, make sets
words2 = set(s2.split())

common = words1 & words2
unique1 = words1 - words2
unique2 = words2 - words1

print("Common words:", common)
print("Words unique to first:", unique1)
print("Words unique to second:", unique2)
```

```python
# Program to perform all set operations on three sets
A = set(map(int, input("Enter elements of set A: ").split()))
B = set(map(int, input("Enter elements of set B: ").split()))
C = set(map(int, input("Enter elements of set C: ").split()))

union_all = A | B | C
inter_all = A & B & C
diff_AB = A - B
diff_BC = B - C
sym_diff_all = A ^ B ^ C                     # symmetric difference across three

print("Union of A, B, C:", union_all)
print("Intersection of A, B, C:", inter_all)
print("A - B:", diff_AB)
print("B - C:", diff_BC)
print("Symmetric difference of A, B, C:", sym_diff_all)
```
