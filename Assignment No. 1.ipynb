<h1>Question Set A</h1>

```python
# Program to calculate average of numbers in a given list
L = [22, 45, 65, 82]
c = s = 0
for l in L:
    c += 1 # Incrementing to count total values in the list
    s += l # Adding the values for every loop
Avg = s / c
print(f"Average = {Avg}")
```

```python
# Program to identify and display duplicate values in a given collection of integer values
T = []
for i in range(6): # Loop to collect 6 numbers from user
    l = int(input("Enter a number: ")) 
    T.append(l) # Add the input number to the list T
D = set() # Create an empty set to store unique values
for t in T: 
        D.add(t) # Add each value to the set D
if len(D) != len(T): # Compare lengths of D and T
    print("Duplicates") # If lengths differ, duplicates exist
else:
    print("All Unique") # If lengths are the same, all values are unique
```

```python
# Program to check if entered value is between 1 and 50 (both inclusive) or not
n = int(input("Enter a number: "))
if 1 <= n <= 50:
    print("Ok") # If n is within the range
else:
    print("Out of range") # If n is outside the range
```

```python
# Program to find sum of each digit on a number
n = int(input("Enter a number: "))
s = 0
while n > 0:
    s += n % 10 # Add the last digit to sum
    n //= 10 # Remove the last digit from the number
print("Sum of digits:", s)
```

```python
# Program to print the Fibonacci series
n = int(input("Enter the number of terms in Fibonacci series: "))
a, b = 0, 1
for i in range(n): 
    print(a, end=' ') # Print the current term
    a, b = b, a + b # Update values of a and b for next term
```

<h1>Question Set B</h1>

```python
# Program that accepts a number from user and prints all prime numbers till that number
n = int(input("Enter a number: "))
for i in range(2, n + 1):
    if i % 2 != 0: # Check if iterated value is divisible by 2
        print(i)
```

```python
# Program that checks if two integer values are same or not
a = int(input("Enter first integer: ")) 
b = int(input("Enter second integer: "))
if a == b: # Check if both values are equal
    print("SAME identity")
else:
    print("DIFFERENT identity")
```

```python
# Program to print a pattern
n = 4
for i in range(n, 0, -1): # Loop from n down to 1
    for j in range(1, i + 1): # Loop from 1 to i
        print(j, end=" ") # Print numbers in the same line
    print()
```

```python
# Program to print reverse of a given number
n = int(input("Enter a number: "))
r = 0
while n > 0:
    d = n % 10 # Get the last digit of n
    r = r * 10 + d # Append digit to the reverse number
    n //= 10 # Remove the last digit from n
print("Reversed Number:", r) # Display the reversed number
```

<h1>Question Set C</h1>

```python
# Program to search an element in a sorted list and display the index of that element if found
L = [12, 25, 36, 47, 58, 69, 70]
e = int(input("Enter element to search: ")) # Input element to search
found = False
for i in range(len(L)):
    if L[i] == e: # Check if the current element matches the search element
        print("Element found at index:", i) # Display index if found
        found = True
        break
if not found:
    print("Element not found") # Display message if element is not found
```
